```python
involved_series = ["Benzene-bridges", "benz-b-imbal", "CDT-BT", "TT-BT"]
transition = "C0S0_C0S1"

fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(15,10))
xlabel_fs = 15
axs[0].set_xlabel("Low", fontsize= xlabel_fs)
axs[1].set_xlabel("Mid", fontsize = xlabel_fs)
axs[2].set_xlabel("High", fontsize = xlabel_fs)
axs[3].set_xlabel("Total", fontsize = xlabel_fs)

max_lim = 0
min_lim = 0

plot_legends = True
for series in involved_series:
    charge_state_1 = transition.split("_")[0]
    charge_state_2 = transition.split("_")[1]
    
    alt_name = charge_state_2 + "_" + charge_state_1
    
    single_c = "#AABD8C"
    c2_c     = "#F39B6D"
    c3_c     = "#3E6990"
    benz_c   = "#381D2A"
    colors = [single_c, c2_c, c3_c, benz_c]
    bridge_labels = ["Single", "Ethylene", "Acetylene", "Benzene"]
    label_loc = [0,1,2]
    label_dev = [0,0.2,0.4,0.6]
    marker_size = 100
    
    #loading data ---------------------------------------------------------------------------
    
    #single bond
    single_serial = "_0_"
    try:
        single = pd.read_csv(Reorg_folder+series+single_serial+transition)
    except:
        try: 
            single = pd.read_csv(Reorg_folder+series+single_serial+alt_name)
        except:
            pass
    
    single_low  = single[single['Frequencies']<500 ]["Reorg Energy"].sum()*cmtoEv
    single_mid  = single[single['Frequencies']<2000][single['Frequencies']>=500]["Reorg Energy"].sum()*cmtoEv
    single_high = single[single['Frequencies']>=2000]["Reorg Energy"].sum()*cmtoEv
    single_tot  = single["Reorg Energy"].sum()*cmtoEv
    
        
    #polyacetylene bond
    c2_serial = "_1_"
    try:
        c2 = pd.read_csv(Reorg_folder+series+c2_serial+transition)
    except:
        try: 
            c2 = pd.read_csv(Reorg_folder+series+c2_serial+alt_name)
        except:
            pass
    
    c2_low  = c2[c2['Frequencies']<500 ]["Reorg Energy"].sum()*cmtoEv
    c2_mid  = c2[c2['Frequencies']<2000][c2['Frequencies']>=500]["Reorg Energy"].sum()*cmtoEv
    c2_high = c2[c2['Frequencies']>=2000]["Reorg Energy"].sum()*cmtoEv
    c2_tot  = c2["Reorg Energy"].sum()*cmtoEv
    
    
    #acetylene bond
    c3_serial = "_2_"
    try:
        c3 = pd.read_csv(Reorg_folder+series+c3_serial+transition)
    except:
        try: 
            c3 = pd.read_csv(Reorg_folder+series+c3_serial+alt_name)
        except:
            pass
        
    c3_low  = c3[c3['Frequencies']<500 ]["Reorg Energy"].sum()*cmtoEv
    c3_mid  = c3[c3['Frequencies']<2000][c3['Frequencies']>=500]["Reorg Energy"].sum()*cmtoEv
    c3_high = c3[c3['Frequencies']>=2000]["Reorg Energy"].sum()*cmtoEv
    c3_tot  = c3["Reorg Energy"].sum()*cmtoEv


    #benzene bridge bond
    benzene_serial = "_3_"
    try:
        benz = pd.read_csv(Reorg_folder+series+benzene_serial+transition)
    except:
        try: 
            benz = pd.read_csv(Reorg_folder+series+benzene_serial+alt_name)
        except:
            pass
        
    benz_low  = benz[benz['Frequencies']<500 ]["Reorg Energy"].sum()*cmtoEv
    benz_mid  = benz[benz['Frequencies']<2000][benz['Frequencies']>=500]["Reorg Energy"].sum()*cmtoEv
    benz_high = benz[benz['Frequencies']>=2000]["Reorg Energy"].sum()*cmtoEv
    benz_tot  = benz["Reorg Energy"].sum()*cmtoEv
    
    
    #processing --------------------------------------------------------------------------------------
    
    low  = [single_low, c2_low, c3_low, benz_low]
    mid  = [single_mid, c2_mid, c3_mid, benz_mid]
    high = [single_high, c2_high, c3_high, benz_high]
    tot = [single_tot, c2_tot, c3_tot, benz_tot]
    
    low_dev  = [(x - np.mean(low))/np.mean(low) for x in low]
    mid_dev  = [(x - np.mean(mid))/np.mean(mid) for x in mid]
    high_dev = [(x - np.mean(high))/np.mean(high) for x in high]
    tot_dev = [(x - np.mean(tot))/np.mean(tot) for x in tot]
    
    #low_dev = [x/max(low) for x in low]
    #mid_dev = [x/max(mid) for x in mid]
    #high_dev = [x/max(high) for x in high]
    
    #plotting data ------------------------------------------------------------------------------------

    if plot_legends:
        for i in range(0,len(low)):
            axs[0].scatter(label_dev[i], low_dev[i], marker='o', color = colors[i], s=marker_size, label = bridge_labels[i])
            axs[1].scatter(label_dev[i], mid_dev[i], marker='o', color = colors[i], s=marker_size, label = bridge_labels[i])
            axs[2].scatter(label_dev[i], high_dev[i], marker='o', color = colors[i], s=marker_size, label = bridge_labels[i])
            axs[3].scatter(label_dev[i], tot_dev[i], marker='o', color = colors[i], s=marker_size, label = bridge_labels[i])
    else:
        for i in range(0,len(low)):
            axs[0].scatter(label_dev[i], low_dev[i], marker='o', color = colors[i], s=marker_size)
            axs[1].scatter(label_dev[i], mid_dev[i], marker='o', color = colors[i], s=marker_size)
            axs[2].scatter(label_dev[i], high_dev[i], marker='o', color = colors[i], s=marker_size)
            axs[3].scatter(label_dev[i], tot_dev[i], marker='o', color = colors[i], s=marker_size)
        
    plot_legends = False
    

    max_series = max(max(low_dev), max(mid_dev), max(high_dev))
    min_series = min(min(low_dev), min(mid_dev), min(high_dev))
    if max_series > max_lim:
        max_lim = max_series
    
    if min_series < min_lim:
        min_lim = min_series

axs[0].legend(fontsize=13)
axs[0].set_xticks([])
axs[1].set_xticks([])
axs[2].set_xticks([])
axs[3].set_xticks([])
    


```
